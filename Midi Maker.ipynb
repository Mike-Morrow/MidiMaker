```python
%matplotlib inline

from midiutil.MidiFile import MIDIFile
import random
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#list out possible note pitches
#keep track of current time
#select note and duration 

#Blues C maj: C, D, D flat, E, G, A

blues_scale = [60, 62, 63, 64, 67, 69]
# blues_scale = [60, 100]
```

```python
mf = MIDIFile(1)     # only 1 track
track = 0   # the only track

time = 0    # start at the beginning
mf.addTrackName(track, time, "Sample Track")
mf.addTempo(track, time, 120)

channel = 0
volume = 100

time = 0    # start at the beginning
```

```python
print(time)
```

```python
time = 0
duration = 0
note_list = []
for i in range(100):
#     pitch = random.randrange(60,70)
    pitch = np.random.choice(a=blues_scale)
    time = time + duration 
    duration = random.randrange(1,3)
    #time = time + last iterations duration
    
#     print(time,duration)
    note_list.append([pitch, duration, time])
    #print(duration)
    mf.addNote(track, channel, pitch, time, duration, volume)
```

```python
track_df = pd.DataFrame(note_list, columns=['Pitch', 'Duration', 'Time'])
```

```python
# write it to disk
with open("output.mid", 'wb') as outf:
    mf.writeFile(outf)
```
